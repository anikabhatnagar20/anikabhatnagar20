```python
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Student Grades</title>
</head>
<body>

<button onclick="start()">Start Adding Students</button>
<p id="output"></p>

<script>
    let studentGrades = {};
    let passingStudents = [];

    function start() {
        let continueAdding = true;

        while (continueAdding) {
            let name = prompt("Enter the student's name:");

            if (name === null) { // If user presses 'cancel' on the prompt
                break;
            }

            let gradeStr = prompt("Enter the grade for " + name + ":");

            if (gradeStr === null) { // If user presses 'cancel' on the prompt
                break;
            }

            let grade = parseFloat(gradeStr);

            // Nested if/else
            if (name && !isNaN(grade)) {
                studentGrades[name] = grade;
            } else {
                alert("Invalid input. Please enter a valid name and grade.");
                continue;
            }

            let more = prompt("Do you want to add more students? (yes/no)").toLowerCase();

            if (more === "no") {
                continueAdding = false;
            }
        }

        processGrades();
    }

    function processGrades() {
        let highestScore = -Infinity;

        for (let grade of Object.values(studentGrades)) {
            if (grade > highestScore) {
                highestScore = grade;
            }
        }

        for (let [name, grade] of Object.entries(studentGrades)) {
            if (grade >= 50) {
                passingStudents.push(name);
            }
        }

        document.getElementById("output").innerText = 
            "Highest score is: " + highestScore + "\n" + 
            "Students who passed: " + passingStudents.join(", ");
    }
```
